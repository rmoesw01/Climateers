```python
# startDates = ["1880-01-01", "1890-01-01", "1900-01-01", "1910-01-01", "1920-01-01", "1930-01-01", "1940-01-01", "1950-01-01", "1960-01-01", "1970-01-01", "1980-01-01", "1990-01-01", "2000-01-01", "2010-01-01", "2020-01-01"]
# endDates = ["1889-12-31", "1899-12-31", "1909-12-31", "1919-12-31", "1929-12-31", "1939-12-31", "1949-12-31", "1959-12-31", "1969-12-31", "1979-12-31", "1989-12-31", "1999-12-31", "2009-12-31", "2019-12-31", "2020-10-24"]

startDates = ["1940-01-01", "1945-01-01", "1950-01-01", "1955-01-01", "1960-01-01", "1965-01-01", "1970-01-01", "1975-01-01", "1980-01-01", "1985-01-01", "1990-01-01", "1995-01-01", "2000-01-01", "2005-01-01", "2010-01-01", "2015-01-01"]
endDates = ["1944-12-31", "1949-12-31", "1954-12-31", "1959-12-31", "1964-12-31", "1969-12-31", "1974-12-31", "1979-12-31", "1984-12-31", "1989-12-31", "1994-12-31", "1999-12-31", "2004-12-31", "2009-12-31", "2014-12-31", "2019-12-31"]

# for j in range(0,len(startDates)):
for j in range(0,8):
    startDate = startDates[j]
    endDate = endDates[j]
    url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/data"
    data = {
        "datasetid":"GSOY",
        "limit": 1000,
        "startdate": startDate,
        "enddate": endDate
    }
    headers = {"token": NOAA_token}

    response = requests.get(url, params=data, headers=headers)
    # print(response)
    response_json = response.json()
    currGSOYdata = pd.json_normalize(response_json["results"])

    numResults = response_json["metadata"]["resultset"]["count"]
    print(numResults)
    for i in range(1, int(numResults/1000)+1):
        currOffset = i * 1000
        url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/data"
        data = {
            "datasetid":"GSOY",
            "limit": 1000,
            "startdate": startDate,
            "enddate": endDate,
            "offset": currOffset
        }
        headers = {"token": NOAA_token}

        response = requests.get(url, params=data, headers=headers)
        print(currOffset, response)
        response_json = response.json()
        currResults = pd.json_normalize(response_json["results"])
        currGSOYdata = currGSOYdata.append(currResults)

    currGSOYdata.to_csv("static/data/NOAA_GSOYdata"+startDates[j][0:4]+"s.csv", index=False)
```
