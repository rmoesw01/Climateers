```python
import pandas as pd
```

```python
wind_df = pd.read_csv("Data/wind-energy-consumption-by-region.csv")
solar_df = pd.read_csv("Data/solar-energy-consumption-by-region.csv")
biofuel_df = pd.read_csv("Data/biofuels-production-by-region.csv")
hydropower_df = pd.read_csv("Data/hydro-consumption-by-region.csv")
```

```python
wind_df
```

```python
test1 = solar_df.groupby(["Entity","Year"]).sum()
type(test1)
```

```python
test2 = wind_df.groupby(["Entity","Year"]).sum()
test2
```

```python
merged_df = test1.merge(test2, how='inner', left_on=["Entity", "Year"], right_on=["Entity","Year"])
merged_df
```

```python
test3 = hydropower_df.groupby(["Entity","Year"]).sum()
test3
```

```python
merged_df_2 = merged_df.merge(test3, how='inner', left_on=["Entity", "Year"], right_on=["Entity","Year"])
merged_df_2
```

```python
test4 = biofuel_df.groupby(["Entity","Year"]).sum()
test4
```

```python
merged_df_3 = merged_df_2.merge(test4, how='outer', left_on=["Entity", "Year"], right_on=["Entity","Year"])
merged_df_3
```

```python
merged_df_3 = merged_df_3.fillna(0)
```

```python
merged_df_3.to_csv("Data/Energy_consume_region.csv")
```
