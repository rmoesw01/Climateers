```python
import pandas as pd
```

```python
# clean_data
glacier=pd.read_csv('Resources/glacier_mass.csv')
co2_level=pd.read_csv('Resources/global_co2_level.csv')
emissions=pd.read_csv('Resources/global_emissions.csv')
temp=pd.read_csv('Resources/global_temp.csv')
sea_level=pd.read_csv('Resources/sea_level_change.csv')
```

```python
print(len(emissions['Total']))
emissions.head()
```

```python
print(temp.Year.min())
temp.head()
```

```python
# set all date ranges the same
glacier=glacier.loc[glacier['Year']>= 1959]
glacier=glacier.loc[glacier['Year']<= 2010]
glacier.head()
```

```python
sea_level=sea_level.loc[sea_level['Time']>= '1959-03-15']
sea_level=sea_level.loc[sea_level['Time']<= '2010-03-15']
sea=sea_level.GMSL.to_list()
```

```python
co2_level=co2_level.loc[co2_level['Year']>= '1959-11-01']
co2_level=co2_level.loc[co2_level['Year']<= '2010-11-01']
co2=co2_level['Annual Increase'].to_list()
```

```python
emissions.head()
```

```python
emissions=emissions.loc[emissions['Year']>= 1959]
emissions=emissions.loc[emissions['Year']<= 2010]
emissions.describe()
```

```python
temp=temp.loc[temp['Year']>= 1959]
temp=temp.loc[temp['Year']<= 2010]
temp.head()
```

```python
gcag_temp=temp.loc[temp['Source']=='GCAG']
gcag_temp=gcag_temp.sort_values("Year")
gcag_temp_val=gcag_temp.Mean.to_list()
gcag_temp_val
```

```python
gis_temp=temp.loc[temp['Source']=='GISTEMP']
gis_temp=gis_temp.sort_values("Year")
gis_temp_val=gis_temp.Mean.to_list()
gis_temp_val
```

```python
model_set=emissions
model_set['gis_temp']=gis_temp_val
model_set['gcag_temp']=gcag_temp_val
model_set['co2_level']=co2
model_set['sea_level']=sea
```

```python
model_set.head()
```

```python
del model_set['Total']
model_set.head()
```
