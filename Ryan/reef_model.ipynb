```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('Resources/reef.csv')
print(data.columns)
data
```

```python
data=data.drop([0])
data
```

```python
clean_reef=data[['depth', 'Average_Bleaching', 'ClimSST',
       'Temperature_Kelvin', 'Temperature_Minimum',
       'Temperature_Maximum', 'Temperature_Kelvin_Standard_Deviation',
        'SSTA', 'SSTA_Standard_Deviation',
       'SSTA_Minimum', 'SSTA_Maximum', 'SSTA_Frequency',
       'SSTA_Frequency_Standard_Deviation', 'SSTA_FrequencyMax',
       'SSTA_FrequencyMean', 'SSTA_DHW', 'SSTA_DHW_Standard_Deviation',
       'SSTA_DHWMax', 'SSTA_DHWMean', 'TSA', 'TSA_Standard_Deviation',
       'TSA_Minimum', 'TSA_Maximum',  'TSA_Frequency',
       'TSA_Frequency_Standard_Deviation', 'TSA_FrequencyMax',
        'TSA_DHW', 'TSA_DHW_Standard_Deviation',
       'TSA_DHWMax'
                ]]
clean_reef
```

```python
clean_reef=clean_reef.apply(pd.to_numeric)
```

```python
clean_reef.isnull().sum().sum()
```

```python
clean_reef=clean_reef.dropna()
```

```python
X=clean_reef.drop(columns=['Temperature_Kelvin', 'Temperature_Minimum',
       'Temperature_Maximum', 'Temperature_Kelvin_Standard_Deviation'])
y=clean_reef.Temperature_Kelvin
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)



print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
```
