```python
atlantic_hurricane_df = pd.read_csv("Data/atlantic_hurricanes.csv")
atlantic_hurricane_df
```

```python
#format date
# https://datatofish.com/integers-datetime-pandas-dataframe/
atlantic_hurricane_df['Date'] = pd.to_datetime(atlantic_hurricane_df['Date'], format='%Y%m%d')
```

```python
#extract year and add as a column
atlantic_hurricane_df['Year'] = atlantic_hurricane_df['Date'].dt.year
atlantic_hurricane_df
```

```python
atlantic_hurricane_df.columns
```

```python
atlantic_hurricane_df["Count"] = 1
atlantic_hurricane_df
```

```python
atlantic_hurricane_df = atlantic_hurricane_df[['Date','Year', 'Status','Count']]
atlantic_hurricane_df
```

```python
years = atlantic_hurricane_df["Year"].unique()
hurricanes = []
tropical_storms = []
tropical_depression = []
for year in years:
    sum_HU = atlantic_hurricane_df.loc[(atlantic_hurricane_df["Status"] == ' HU') & 
                                    (atlantic_hurricane_df["Year"] == year),:]
    hurricanes.append(sum_HU.shape[0])
    sum_TS = atlantic_hurricane_df.loc[(atlantic_hurricane_df["Status"] == ' TS') & 
                                    (atlantic_hurricane_df["Year"] == year),:]
    tropical_storms.append(sum_TS.shape[0])
    sum_TD = atlantic_hurricane_df.loc[(atlantic_hurricane_df["Status"] == ' TD') & 
                                    (atlantic_hurricane_df["Year"] == year),:]
    tropical_depression.append(sum_TD.shape[0])
```

```python
#zip together above lists
atlantic_ocean_data = list(zip(years, hurricanes, tropical_storms,tropical_depression))
#make a dataframe from zipped lists and rename columns
atlantic_ocean_data = pd.DataFrame(atlantic_ocean_data)
atlantic_ocean_data = atlantic_ocean_data.rename(columns = {0:"Year", 1:"Hurricanes", 2:"Tropical_Storms", 3:"Tropical_Depression"})
# atlantic_ocean_data = atlantic_ocean_data[:62]
atlantic_ocean_data = atlantic_ocean_data[29:160]
atlantic_ocean_data
```

```python
gw_data_condensed_hurricanes = pd.merge(gw_data_condensed, atlantic_ocean_data, on = "Year", how = "inner")
gw_data_condensed_hurricanes
```

```python
X=gw_data_condensed_hurricanes[['Gas Flaring','Liquid Fuel','Solid Fuel','Cement','Gas Fuel','GCAG','Annual_Sea_anomaly']]
y=gw_data_condensed_hurricanes.Tropical_Depression
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#modeling Tropical Depressions
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
