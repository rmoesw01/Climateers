```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read in the data
hurricane_atlantic_df = pd.read_csv("Data/atlantic_hurricanes.csv")
```

```python
#display data
hurricane_atlantic_df
```

```python
#show columns
hurricane_atlantic_df.columns
```

```python
#format date
# https://datatofish.com/integers-datetime-pandas-dataframe/
hurricane_atlantic_df['Date'] = pd.to_datetime(hurricane_atlantic_df['Date'], format='%Y%m%d')
```

```python
#extract year and add as a column
hurricane_atlantic_df['Year'] = hurricane_atlantic_df['Date'].dt.year
```

```python
#display dataframe
hurricane_atlantic_df
```

```python
#groupby year and count
hurricane_atlantic_df.groupby(["Year"]).count()
```

```python
#groupby event and count
hurricane_atlantic_df.groupby(["Status"]).count()
```

```python
hurricane_atlantic_df["Year"].unique()
x = hurricane_atlantic_df["Year"].unique()
y = hurricane_atlantic_df.groupby(["Year"])["ID"].count()
plt.plot(x,y)
```

```python
hurricanes_only = hurricane_atlantic_df.loc[hurricane_atlantic_df["Status"]==" HU",:]
hurricanes_only
```

```python
#hurricanes only
x_hurricanes = hurricanes_only["Year"].unique()
y_hurricanes = hurricanes_only.groupby(["Year"])["ID"].count()
plt.plot(x_hurricanes,y_hurricanes)
```

```python
T_storms = hurricane_atlantic_df.loc[hurricane_atlantic_df["Status"]==" TS",:]
T_storms
```

```python
#Tropical storms only
x_T_storms = T_storms["Year"].unique()
y_T_storms= T_storms.groupby(["Year"])["ID"].count()
plt.plot(x_T_storms,y_T_storms)
```
