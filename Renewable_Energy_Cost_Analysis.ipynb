```python
import pandas as pd
```

```python
renewable_invest_df = pd.read_csv("Data/investment-in-renewable-energy-by-technology.csv")
renewable_invest_df
```

```python
renewable_gen_df = pd.read_csv("Data/modern-renewable-prod.csv")
renewable_gen_df = renewable_gen_df.loc[renewable_gen_df["Entity"]=="World",:]
renewable_gen_df = renewable_gen_df[39:52]
renewable_gen_df
```

```python
invest_gen_df = pd.merge(renewable_gen_df,renewable_invest_df, on = "Year", how = "inner" )
invest_gen_df
```

```python
invest_gen_df["Solar gen/cost"] = invest_gen_df["Electricity from solar (TWh)"]/invest_gen_df["Solar Energy"]
invest_gen_df
```

```python
invest_gen_df["Wind gen/cost"] = invest_gen_df["Electricity from wind (TWh)"]/invest_gen_df["Wind Energy"]
invest_gen_df["Hydro gen/cost"] = invest_gen_df["Electricity from hydro (TWh)"]/50
invest_gen_df
```
