```python
import pandas as pd
```

```python
global_warming_df = pd.read_csv("Data/ENT.csv")
global_warming_df
```

```python
sea_temps = pd.read_csv("Data/Sea_temperature.csv")
sea_temps
```

```python

sea_temps = sea_temps.rename(columns={"Annual anomaly":"Annual_Sea_anomaly"})
sea_temps = sea_temps[["Year","Annual_Sea_anomaly"]]
sea_temps
```

```python
gw_data_condensed = pd.merge(global_warming_df, sea_temps, on = "Year", how = "outer")
gw_data_condensed.head(30)
```

```python
del gw_data_condensed['Unnamed: 0']
# del gw_data_condensed['Year']
del gw_data_condensed['Total']
```

```python
gw_data_condensed
```

```python
sea_level = pd.read_csv("Data/sea_level_change.csv")
sea_level
```

```python
#format date
# https://datatofish.com/integers-datetime-pandas-dataframe/
sea_level['Time'] = pd.to_datetime(sea_level['Time'], format='%Y-%m-%d')
```

```python
sea_level['Year'] = sea_level['Time'].dt.year
sea_level
```

```python
sea_level = sea_level[['Year', 'GMSL']]
sea_level
```

```python
gw_data_condensed_intermediate = pd.merge(gw_data_condensed, sea_level, on = "Year", how = "outer")
gw_data_condensed_intermediate
```

```python
co2_level = pd.read_csv("Data/global_co2_level.csv")
co2_level.head()
```

```python
co2_level = co2_level.rename(columns = {"Year":"Date", "Annual Increase":"CO2_Increase"})
co2_level.head()
```

```python
#format date
# https://datatofish.com/integers-datetime-pandas-dataframe/
co2_level['Date'] = pd.to_datetime(co2_level['Date'], format='%m/%d/%Y')
co2_level.head()
```

```python
co2_level['Year'] = co2_level['Date'].dt.year
co2_level.head()
```

```python
co2_level = co2_level[['Year', 'CO2_Increase']]
co2_level.head()
```

```python
gw_data_condensed_intermediate_2 = pd.merge(gw_data_condensed_intermediate, co2_level, on = "Year", how = "outer")
gw_data_condensed_intermediate_2
```

```python
gw_data_condensed_intermediate_2 = gw_data_condensed_intermediate_2.rename(columns = {"Annual_Sea_anomaly":"Sea_anomaly_temp"})
```

```python
gw_data_condensed_intermediate_2
```

```python
gw_data_condensed_intermediate_2.to_csv('Data/All_parameters.csv')
```
