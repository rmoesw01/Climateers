```python
#import dependencies
import pandas as pd
```

```python
#read in global warming parameters
all_parameters = pd.read_csv("../general/All_parameters.csv")
all_parameters
```

```python
#read in tornado data
tornado_df = pd.read_csv("1950-2018alltornadoes.csv")
tornado_df
```

```python
#delete unneeded columns
del tornado_df['om']
del tornado_df['mo']
del tornado_df['dy']
del tornado_df['date']
del tornado_df['time']
```

```python
#show dataframe
tornado_df
```

```python
#update column names
tornado_df_clean = tornado_df.rename(columns = {"yr":"Year","st": "State", "inj":"Injuries", "fat":"Fatalities","mag":"Magnitude",
                                                "loss":"Property_Loss","closs":"Crop_loss","slat":"Lat_Start",
                                                "slon":"Long_Start","elat":"Lat_End","elon":"Long_End", "len":"Length_(mi)", "wid":"Width_(yds)"})
```

```python
#show new df
tornado_df_clean
```

```python
#remove unneeded columns
tornado_df_clean_reduced = tornado_df_clean[["Year", "Magnitude","State","Injuries","Fatalities",  "Lat_Start","Long_Start", "Lat_End","Long_End", "Length_(mi)","Width_(yds)"]]
```

```python
#add a count column with value 1 for all rows
tornado_df_clean_reduced["Count"]=1
tornado_df_clean_reduced
```

```python
# count number of each magnitude for each year and add to individual list, as well as count for total tornadoes each year
years = tornado_df_clean_reduced["Year"].unique()
Mag_0 = []
Mag_1 = []
Mag_2 = []
Mag_3 = []
Mag_4 = []
Mag_5 = []
tornado_total = []

for year in years:
    total = tornado_df_clean_reduced.loc[(tornado_df_clean_reduced["Year"] == year),:]
    tornado_total.append(total.shape[0])
    sum_0 = tornado_df_clean_reduced.loc[(tornado_df_clean_reduced["Magnitude"] == 0) & 
                                    (tornado_df_clean_reduced["Year"] == year),:]
    Mag_0.append(sum_0.shape[0])
    sum_1 = tornado_df_clean_reduced.loc[(tornado_df_clean_reduced["Magnitude"] == 1) & 
                                    (tornado_df_clean_reduced["Year"] == year),:]
    Mag_1.append(sum_1.shape[0])
    sum_2 = tornado_df_clean_reduced.loc[(tornado_df_clean_reduced["Magnitude"] == 2) & 
                                    (tornado_df_clean_reduced["Year"] == year),:]
    Mag_2.append(sum_2.shape[0])
    sum_3 = tornado_df_clean_reduced.loc[(tornado_df_clean_reduced["Magnitude"] == 3) & 
                                    (tornado_df_clean_reduced["Year"] == year),:]
    Mag_3.append(sum_3.shape[0])
    sum_4 = tornado_df_clean_reduced.loc[(tornado_df_clean_reduced["Magnitude"] == 4) & 
                                    (tornado_df_clean_reduced["Year"] == year),:]
    Mag_4.append(sum_4.shape[0])
    sum_5 = tornado_df_clean_reduced.loc[(tornado_df_clean_reduced["Magnitude"] == 5) & 
                                    (tornado_df_clean_reduced["Year"] == year),:]
    Mag_5.append(sum_5.shape[0])
```

```python
#zip together above lists
tornado_data = list(zip(years, Mag_0, Mag_1,Mag_2,Mag_3,Mag_4,Mag_5,tornado_total))
#make a dataframe from zipped lists and rename columns
tornado_data = pd.DataFrame(tornado_data)
tornado_data = tornado_data.rename(columns = {0:"Year", 1:"Magnitude_0", 2:"Magnitude_1", 3:"Magnitude_2", 4: "Magnitude_3",5:"Magnitude_4", 6:"Magnitude_5",7:"Tornado_total"})
# tornado_data = atlantic_ocean_data[29:160]
tornado_data.tail(30)
```

```python
#merge tornado df with global warming df and use only rows with all data available (no NAN entries)
gw_data_tornado = pd.merge(all_parameters, tornado_data, on = "Year", how = "inner")
gw_data_tornado_condensed = gw_data_tornado[9:60]
gw_data_tornado_condensed
```

```python
#show columns
gw_data_tornado_condensed.columns
```

```python
#select inputs for total tornado model
X = gw_data_tornado_condensed[['Gas Fuel',   'Cement',
      'GIS_TEMP',  'population','CO2_Increase',
         'CSIRO Adjusted Sea Level'
       ]]
#set target
y=gw_data_tornado_condensed.Tornado_total
```

```python
#split data into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
#import linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#modeling total tornadoes
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
#get coeffiecent and intercept for total tornado model
model.coef_
model.intercept_
```

```python
#bring in projected values for global warming factors (from linear regression function)
projected_values = pd.read_csv("../Output_Data/Forecasted_global_warming_values_for_models.csv")
projected_values
```

```python
#projecting number of tornadoes
projected_num_tornadoes = []
for i in range(len(projected_values["Year"])):
    proj_tornado_tot = ((model.coef_[0])*(projected_values["Gas Fuel"][i]) + (model.coef_[1])*(projected_values["Cement"][i])\
    + (model.coef_[2])*(projected_values["GIS_TEMP"][i]) + (model.coef_[3])*(projected_values["population"][i]) \
    + (model.coef_[4])*(projected_values["CO2_Increase"][i]) +(model.coef_[5])*(projected_values["CSIRO Adjusted Sea Level"][i]))\
    +model.intercept_
    projected_num_tornadoes.append(proj_tornado_tot)

projected_num_tornadoes
```

```python
#projecting mag 0 tornadoes
projected_num_mag_0 = []
for i in range(len(projected_values["Year"])):
    proj_mag_0_tot = ((model_mag_0.coef_[0])*(projected_values["Gas Fuel"][i]) + (model_mag_0.coef_[1])*(projected_values["Liquid Fuel"][i])\
    + (model_mag_0.coef_[2])*(projected_values["Cement"][i]) + (model_mag_0.coef_[3])*(projected_values["GIS_TEMP"][i]) \
    + (model_mag_0.coef_[4])*(projected_values["population"][i]) +(model_mag_0.coef_[5])*(projected_values["CO2_Increase"][i]))\
    +model_mag_0.intercept_
    projected_num_mag_0.append(proj_mag_0_tot)

projected_num_mag_0
```

```python
#projecting mag 2 tornadoes
projected_num_mag_2 = []
for i in range(len(projected_values["Year"])):
    proj_mag_2_tot = ((model_mag_2.coef_[0])*(projected_values["Gas Fuel"][i]) + (model_mag_2.coef_[1])*(projected_values["Liquid Fuel"][i])\
    + (model_mag_2.coef_[2])*(projected_values["Cement"][i]) + (model_mag_2.coef_[3])*(projected_values["GIS_TEMP"][i]) \
    + (model_mag_2.coef_[4])*(projected_values["population"][i]) +(model_mag_2.coef_[5])*(projected_values["Sea_anomaly_temp"][i]))\
    + model_mag_2.intercept_
    projected_num_mag_2.append(proj_mag_2_tot)

projected_num_mag_2
```

```python
#zip together above lists
projected_tornado_info = list(zip(projected_values["Year"], projected_num_mag_0,projected_num_mag_2, projected_num_tornadoes))
#make a dataframe from zipped lists and rename columns
projected_tornado_info = pd.DataFrame(projected_tornado_info)
#rename columns
projected_tornado_info = projected_tornado_info.rename(columns = {0:"Year", 1:"Magnitude_0", 2:"Magnitude_2", 3:"Tornado_total"})
projected_tornado_info = projected_tornado_info.round()
```

```python
#select which colmns to show
tornado_data_reduced_columns = tornado_data[["Year", "Magnitude_0", "Magnitude_2", "Tornado_total"]]
tornado_data_reduced_columns = tornado_data_reduced_columns[:61]
```

```python
#combine historical tornado data with projected data
current_future_tornadoes_df = pd.concat([tornado_data_reduced_columns,projected_tornado_info], axis =0)
current_future_tornadoes_df.head(30)
```

```python
#create year list to be used, as well as alpha and dashlength needed for visualization
year_start = 1940
alpha = []
dashLength = []
years_reduce = []
for i in range(26):
    year_start = year_start+10
    years_reduce.append(year_start)
    if year_start < 2020:
        alpha_value = 1
        dash_value = 0
    else:
        alpha_value = 0.4
        dash_value = 8
    alpha.append(alpha_value)
    dashLength.append(dash_value)
years_reduce
```

```python
#select every 10 years from full tornado df to show for visualization
current_future_tornadoes_df_reduce = current_future_tornadoes_df.loc[current_future_tornadoes_df["Year"].isin(years_reduce),:]
current_future_tornadoes_df_reduce
```

```python
#create alpha and dashlength columns
current_future_tornadoes_df_reduce["alpha"] = alpha
current_future_tornadoes_df_reduce["dashLength"] = dashLength
current_future_tornadoes_df_reduce
```

```python
#output final csv for visualization
current_future_tornadoes_df_reduce.to_csv("../Output_Data/Projected_tornadoes_reduced.csv", index = False)
```
