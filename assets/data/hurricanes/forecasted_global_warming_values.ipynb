```python
# Dependencies 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import scipy.stats as sts
```

```python
# functions to project global warming values for the future
def forecast_var(x,y):
    forecast_y=[]
    future_x=[]
    x=x[-20:]
    y=y[-20:]
    slope, intercept, rvalue, pvalue, stderr= linregress(x,y)
    start_x=x[-1]
    start_y=y[-1]
    for j in range(2201-start_x):
        j+=1
        new_x=start_x+j
        new_y=slope*(new_x)+intercept
        forecast_y.append(new_y)
        future_x.append(new_x)
    return(future_x, forecast_y)

def forecast_df(df):
    columns=df.columns.to_list()
    proj_df=pd.DataFrame(columns=columns)
    for j in columns:
        if (j=='Year'):
            x=df.Year.to_list()
        else:
            y=df[f'{j}'].to_list()
            future_x, forecast_y=forecast_var(x,y)
            proj_df[f'{j}']=forecast_y
            proj_df['Year']=future_x
    return proj_df
```

```python
#read in the historical global warming data up to the year 2010
global_warming_parameters = pd.read_csv("../general/All_parameters.csv")
global_warming_parameters = global_warming_parameters[:131]
```

```python
# call function to project future values using linear regression
forecasted_values_df = forecast_df(global_warming_parameters)
```
