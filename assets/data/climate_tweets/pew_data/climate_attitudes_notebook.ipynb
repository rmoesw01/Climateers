## american climate attitudes

```python
#dependencies
import pandas as pd

#module to convert .sav to .csv
#https://ofajardo.github.io/pyreadstat_documentation/_build/html/index.html
import pyreadstat

#import file for US-only survey
US_df, US_meta = pyreadstat.read_sav ('US_survey2019.sav')

US_df
```

```python
#get metadata for column name descriptions
US_meta.column_names_to_labels
```

```python
#drop non-relevant columns; see file called "reading-ATP-data.pdf" for more info on how to read demographic data

clean_df = US_df[['INTERVIEW_START_W55', #date
               'ENVIR8_e_W55', #how much do you think the fed govt is doing to reduce effects of climate change
               'EN7_W55', #how much do you think human activity contributes to climate change
               'F_AGECAT', #age category
               'F_SEX', #sex
               'F_EDUCCAT', #education 1
               'F_EDUCCAT2', #education 2
               'F_PARTYSUM_FINAL']] #party summary

clean_df.to_csv ('pew_csvs/full_data.csv', index = False, header = True)

clean_df
```

```python
#create list of demographic columns for looping
demographics = clean_df.columns[3:]

demographics
```

```python
#loop through demographics, output csvs for each group & each question

for item in demographics:
    q1_df = pd.DataFrame (clean_df.groupby(item)['ENVIR8_e_W55'].value_counts())
    q1_df2 = q1_df.rename (columns = ({'ENVIR8_e_W55': 'count'}))
    q1_df2.to_csv (f'pew_csvs/q1_{item}.csv', index = True, header = True)
    
    q2_df = pd.DataFrame (clean_df.groupby(item)['EN7_W55'].value_counts())
    q2_df2 = q2_df.rename (columns = ({'EN7_W55': 'count'}))
    q2_df2.to_csv (f'pew_csvs/q2_{item}.csv', index = True, header = True)

#view example df
q1_df2
```

## global attitudes

```python
#import file for worldwide survey
global_df, global_meta = pyreadstat.read_sav ('global_survey2018.sav')

global_df
```

```python
#get metadata for column name descriptions
global_meta.column_names_to_labels
```

```python
#keep only question on climate & country code
clean_global_df = global_df[['COUNTRY', 'intthreat_climatechange']]

#drop responses w/ 'don't know' or 'refused'
clean_global_df = clean_global_df[(clean_global_df.intthreat_climatechange != 8.0) &
                                  (clean_global_df.intthreat_climatechange != 9.0)]

#check length
len(clean_global_df)
```

```python
#put country name for country code (27 total)
countries = ['Argentina', 
             'Australia',
             'Brazil',
             'Canada',
             'France',
             'Germany',
             'Greece',
             'Hungary',
             'India',
             'Indonesia',
             'Israel',
             'Italy',
             'Japan',
             'Kenya',
             'Mexico',
             'Netherlands',
             'Nigeria',
             'Philippines',
             'Poland',
             'Russia',
             'South Africa',
             'South Korea',
             'Spain',
             'Sweden',
             'Tunisia',
             'United Kingdom',
             'United States']

clean_global_df['country_name'] = ''

#loop through df to populate country name for each row
for index, row in clean_global_df.iterrows():
    number = int(row['COUNTRY']) - 1
    countryname = countries[number]
    clean_global_df.loc [index, 'country_name'] = countryname
    
clean_global_df
```

```python
#group responses by country, output to csv
country_df = pd.DataFrame (clean_global_df.groupby('country_name')['intthreat_climatechange'].value_counts())
country_df2 = country_df.rename (columns = {'intthreat_climatechange':'count'})

country_df2.to_csv (f'pew_csvs/global_survey.csv', index = True, header = True)

country_df2
```
