```python
import pandas as pd
```

```python
combined_energy_df = pd.read_csv("Data/modern-renewable-prod.csv")
combined_energy_df["Entity"].unique()
```

```python
na_df = combined_energy_df.loc[combined_energy_df["Entity"] == "North America", :]
na_df.head()
```

```python
africa_df = combined_energy_df.loc[combined_energy_df["Entity"] == "Africa", :]
africa_df.head()
```

```python
europe_df = combined_energy_df.loc[combined_energy_df["Entity"] == "Europe", :]
europe_df.head()
```

```python
world_df = combined_energy_df.loc[combined_energy_df["Entity"] == "World", :]
world_df.head()
```

```python
china_df = combined_energy_df.loc[combined_energy_df["Entity"] == "China", :]
china_df.head()
```

```python
brazil_df = combined_energy_df.loc[combined_energy_df["Entity"] == "Brazil", :]
brazil_df.head()
```

```python
combined_energy_df = pd.concat([na_df,africa_df, europe_df, world_df, china_df, brazil_df ], axis = 0)
```

```python
del combined_energy_df["Code"]
combined_energy_df
```

```python
combined_energy_df.to_csv("Output_Data/Combined_Energy_by_Region.csv", index = False)
```
