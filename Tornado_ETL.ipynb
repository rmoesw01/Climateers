```python
import pandas as pd
```

```python
tornado_df = pd.read_csv("Data/1950-2018alltornadoes.csv")
tornado_df
```

```python
tornado_df_clean = tornado_df.rename(columns = {"st": "State", "inj":"Injuries", "fat":"Fatalities","mag":"Magnitude",
                                                "loss":"Property_Loss","closs":"Crop_loss","slat":"Lat_Start",
                                                "slon":"Long_Start","elat":"Lat_End","elon":"Long_End", "len":"Length_(mi)", "wid":"Width_(yds)"})
```

```python
tornado_df_clean.columns
```

```python
tornado_df_clean_reduced = tornado_df_clean[["yr","date", "Magnitude","State","Injuries","Fatalities", "Property_Loss",
                                            "Crop_loss", "Lat_Start","Long_Start", "Lat_End","Long_End", "Length_(mi)","Width_(yds)"]]
```

```python
tornado_df_clean_reduced
```

```python
tornado_df_clean_reduced.groupby(["Magnitude"]).count()
```

```python
year = tornado_df_clean_reduced.groupby(["yr"]).count()
```
