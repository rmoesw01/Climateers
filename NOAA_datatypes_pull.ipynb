```python
import requests
import pandas as pd
import json
from config import NOAA_token
```

```python
url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/datatypes"
data = {
    "limit": 1000
}
headers = {"token": NOAA_token}

response = requests.get(url, params=data, headers=headers)
response_json = response.json()
datatypes = pd.json_normalize(response_json["results"])
datatypes
```

```python
url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/datatypes"
data = {
    "limit": 1000,
    "offset":1000
}
headers = {"token": NOAA_token}

response = requests.get(url, params=data, headers=headers)
response_json = response.json()
datatypes2 = pd.json_normalize(response_json["results"])
datatypes = datatypes.append(datatypes2)
datatypes
```

```python
datatypes['mindate'] = pd.to_datetime(datatypes['mindate'])
datatypes['maxdate'] = pd.to_datetime(datatypes['maxdate'])
```

```python
filtDatatypes = datatypes.loc[(datatypes["mindate"] <= "1880-01-01") & (datatypes["maxdate"] >= "2010-01-01")]
filtDatatypes
```

```python
for row in filtDatatypes["name"]:
    print(row)
```
