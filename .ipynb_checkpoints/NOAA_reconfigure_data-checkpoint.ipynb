```python
import requests
import pandas as pd
import json
from config import NOAA_token
```

```python
def GSOYdf_to_dict (currDF):
    GSOYdata_dict = {}

    for row in currDF.iterrows():
        if row[1].date in GSOYdata_dict:
            if row[1].station in GSOYdata_dict[row[1].date]:
                GSOYdata_dict[row[1].date][row[1].station][row[1].datatype] = row[1].value
            else:
                GSOYdata_dict[row[1].date][row[1].station] = {row[1].datatype: row[1].value}
        else:
            GSOYdata_dict[row[1].date] = {row[1].station: {row[1].datatype: row[1].value}}
#     GSOYdata_dict
```

```python
decades = ["1880","1890","1900","1910","1920","1930","1940","1950","1960","1970","1980","1990","2000","2010","2020"]
for i in range(0,15)
    df = pd.read_csv("static/data/NOAA_GSOYdata"+decades[i]+"s.csv")
    GSOYdf_to_dict(df)
```

```python
# GSOY_data_df = pd.DataFrame.from_dict(GSOYdata_dict)
GSOY_data_df = pd.DataFrame.from_dict({(i,j): GSOYdata_dict[i][j] 
                           for i in GSOYdata_dict.keys() 
                           for j in GSOYdata_dict[i].keys()},
                       orient='index')
GSOY_data_df
```

```python
modified = GSOY_data_df.reset_index()
modified
```

```python
modified = modified.rename(columns={"level_0": "date", "level_1": "station"})
modified
```

```python
modified.to_csv("static/data/NOAA_GSOYsummary.csv", index=False)
```
