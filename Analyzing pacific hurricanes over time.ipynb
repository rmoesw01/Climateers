```python
pacific_hurricane_df = pacific_hurricane_df[['Date','Year', 'Status','Count']]
```

```python
#https://www.quora.com/How-do-I-merge-two-rows-into-one-row-by-an-index-using-pandas
new_df = pacific_hurricane_df.groupby(['Year','Status']).agg('sum') 
new_df
```

```python
years = pacific_hurricane_df["Year"].unique()
status = pacific_hurricane_df["Status"].unique()
```

```python
hurricanes = []
tropical_storms = []
tropical_depression = []
for year in years:
    sum_HU = pacific_hurricane_df.loc[(pacific_hurricane_df["Status"] == ' HU') & 
                                    (pacific_hurricane_df["Year"] == year),:]
    hurricanes.append(sum_HU.shape[0])
    sum_TS = pacific_hurricane_df.loc[(pacific_hurricane_df["Status"] == ' TS') & 
                                    (pacific_hurricane_df["Year"] == year),:]
    tropical_storms.append(sum_TS.shape[0])
    sum_TD = pacific_hurricane_df.loc[(pacific_hurricane_df["Status"] == ' TD') & 
                                    (pacific_hurricane_df["Year"] == year),:]
    tropical_depression.append(sum_TD.shape[0])
        
```

```python
#zip together above lists
pacific_ocean_data = list(zip(years, hurricanes, tropical_storms,tropical_depression))
#make a dataframe from zipped lists and rename columns
pacific_ocean_data = pd.DataFrame(pacific_ocean_data)
pacific_ocean_data = pacific_ocean_data.rename(columns = {0:"Year", 1:"Hurricanes", 2:"Tropical_Storms", 3:"Tropical_Depression"})
pacific_ocean_data = pacific_ocean_data[:62]
pacific_ocean_data
```

```python
gw_data_condensed = gw_data_condensed[69:]
gw_data_condensed
```

```python
gw_data_condensed_hurricanes = pd.merge(gw_data_condensed, pacific_ocean_data, on = "Year", how = "inner")
gw_data_condensed_hurricanes
```

```python
X=gw_data_condensed_hurricanes[['Gas Fuel','Liquid Fuel','Solid Fuel','Gas Flaring','GISTEMP','Annual_Sea_anomaly','population']]
y=gw_data_condensed_hurricanes.Tropical_Depression
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#modeling Tropical Depressions
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
