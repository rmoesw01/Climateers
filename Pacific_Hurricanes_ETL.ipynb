```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pacific_hurricanes_df = pd.read_csv("Data/pacific_hurricanes.csv")
```

```python
pacific_hurricanes_df
```

```python
pacific_hurricanes_df.columns
```

```python
#format date
# https://datatofish.com/integers-datetime-pandas-dataframe/
pacific_hurricanes_df['Date'] = pd.to_datetime(pacific_hurricanes_df['Date'], format='%Y%m%d')
```

```python
#extract year and add as a column
pacific_hurricanes_df['Year'] = pacific_hurricanes_df['Date'].dt.year
```

```python
#groupby year and count
pacific_hurricanes_df.groupby(["Year"]).count()
```

```python
#groupby event and count
pacific_hurricanes_df.groupby(["Status"]).count()
```

```python
pacific_hurricanes_df["Year"].unique()
x = pacific_hurricanes_df["Year"].unique()
y = pacific_hurricanes_df.groupby(["Year"])["ID"].count()
plt.plot(x,y)
```

```python
#groupby year and count
pacific_hurricanes_df.groupby(["Status"]).count()
```

```python
pacific_hurricanes_df["Status"].unique()
x_status = pacific_hurricanes_df["Status"].unique()
y_status = pacific_hurricanes_df.groupby(["Status"])["ID"].count()
plt.plot(x_status,y_status)
```

```python
hurricanes = pacific_hurricanes_df.loc[pacific_hurricanes_df["Status"]==" HU",:]
```

```python
hurricanes
```

```python
#hurricanes only
x_hurricanes = hurricanes["Year"].unique()
y_hurricanes = hurricanes.groupby(["Year"])["ID"].count()
plt.plot(x_hurricanes,y_hurricanes)
```

```python
T_storms = pacific_hurricanes_df.loc[pacific_hurricanes_df["Status"]==" TS",:]
T_storms
```

```python
#tropical storms only
x_T_storms= T_storms["Year"].unique()
y_T_storms = T_storms.groupby(["Year"])["ID"].count()
plt.plot(x_T_storms,y_T_storms)
```
