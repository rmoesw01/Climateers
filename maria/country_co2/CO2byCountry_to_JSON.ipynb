```python
country_df = pd.read_csv("country_codes.csv")
country_df
```

```python
country_df2 = country_df.dropna()
country_df2
```

```python
country_df2.loc [country_df2['Country'] == 'United States of America', 'Country'] = 'United States'
```

```python
issues = []

for x in range(len(co2_2018_df['Entity'])):
    country = co2_2018_df.iloc[x, 0]
#     print (country)
    
    try:
        (country_df2.loc [country_df2['Country'] == country, 'Country']).to_list()[0]
        
    except:
        print (f'{country} not found')
        issues.append (country)
    
print (issues)
```

```python
misfits = ['Iran (Islamic Republic of)', 'Russian Federation', 'Korea, Republic of', 'Syrian Arab Republic', 'Taiwan, Province of China', 'United Kingdom of Great Britain and Northern Ireland', 'Venezuela (Bolivarian Republic of)', 'Viet Nam']
new_names = ['Iran', 'Russia', 'South Korea', 'Syria', 'Taiwan', 'United Kingdom', 'Venezuela', 'Vietnam']


for x in range(len(misfits)):
    print (misfits[x], new_names[x])
    country_df2.loc [country_df2['Country'] == misfits[x], 'Country'] = new_names[x]
    
    try:
        country_df2.loc [country_df2['Country'] == new_names[x], 'Country']
        print ('hooray')
    except:
        print ('you missed a spot')
```

```python
features = []

for x in range(len(co2_2018_df['Entity'])):
    country = co2_2018_df.iloc[x, 0]
    emissions = co2_2018_df.iloc[x, 3]
#     print (country)
    
    try:
        code = (country_df2.loc [country_df2['Country'] == country, 'Code']).to_list()[0]
        features.append(
            {
                "id": code,
                "value": emissions
            }
        )
        
    except:
        pass
        
# print (features)

with open("country_co2.json", "w") as f:
    f.write("%s" % features)
```
