```python
import requests
import pandas as pd
import json
from config import NOAA_token
```

```python
url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/stations"
data = {
    "limit": 1000
}
headers = {"token": NOAA_token}

response = requests.get(url, params=data, headers=headers)
response_json = response.json()
stations = pd.json_normalize(response_json["results"])
numStations = response_json["metadata"]["resultset"]["count"]
stations
```

```python
for i in range(1, int(numStations/1000)+1):
    currOffset = i * 1000
    url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/stations"
    data = {
        "limit": 1000,
        "offset": currOffset
    }
    headers = {"token": NOAA_token}

    response = requests.get(url, params=data, headers=headers)
    print(currOffset, response)
    response_json = response.json()
    stations2 = pd.json_normalize(response_json["results"])
    stations = stations.append(stations2)
```

```python
stations.to_csv("static/data/NOAA_stations.csv", index=False)
```
