```python
station = "GHCND:AGE00147708"
startDate = "2010-01-01"
endDate = "2010-12-31"
# for station in filtStations["id"]:
url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/data"
data = {
    "datasetid":"GHCND",
    "limit": 1000,
    "stationid": station,
    "startdate": startDate,
    "enddate": endDate
}
headers = {"token": NOAA_token}

response = requests.get(url, params=data, headers=headers)
print(response)
response_json = response.json()
station1Data = pd.json_normalize(response_json["results"])
station1Data
```
