```python
import pandas as pd
```

```python
global_warming_parameters = pd.read_csv("Data/All_parameters.csv")
global_warming_parameters.tail(20)
```

```python
global_warming_parameters.columns
```

```python
#showing rows with all data available
global_warming_parameters_reduced = global_warming_parameters[79:131]
```

```python
#predicting sea temp
X_sea_temp=global_warming_parameters_reduced[['GIS_TEMP', 'population','Glacier_mass',
         'CSIRO Adjusted Sea Level']]
y_sea_temp=global_warming_parameters_reduced.Sea_anomaly_temp
```

```python
from sklearn.model_selection import train_test_split
X_train_sea, X_test_sea, y_train_sea, y_test_sea = train_test_split(X_sea_temp, y_sea_temp, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model_sea = LinearRegression()
```

```python
#modeling sea temps
model_sea.fit(X_train_sea, y_train_sea)
training_score_sea = model_sea.score(X_train_sea, y_train_sea)
testing_score_sea = model_sea.score(X_test_sea, y_test_sea)

print(f"Training Score: {training_score_sea}")
print(f"Testing Score: {testing_score_sea}")
```

```python
#get model coefficients

model_sea.coef_
model_sea.intercept_
```

```python
projected_values = pd.read_csv("Data/Forecasted_global_warming_values_for_models.csv")
projected_values
```

```python
projected_sea_temps = []
for i in range(len(projected_values["Year"])):
    sea_tmp = ((model_sea.coef_[0])*(projected_values["GIS_TEMP"][i]) + (model_sea.coef_[1])*(projected_values["population"][i])\
    + (model_sea.coef_[2])*(projected_values["Glacier_mass"][i]) + (model_sea.coef_[3])*(projected_values["CSIRO Adjusted Sea Level"][i]))\
    + model_sea.intercept_
    projected_sea_temps.append(sea_tmp)

projected_sea_temps
```

```python
#zip together above lists
projected_sea_tmp_info = list(zip(projected_values["Year"],projected_sea_temps, projected_values["Sea_anomaly_temp"]))
#make a dataframe from zipped lists and rename columns
projected_sea_tmp_info = pd.DataFrame(projected_sea_tmp_info)
#rename columns
projected_sea_tmp_info = projected_sea_tmp_info.rename(columns = {0:"Year", 1:"Projected_Sea_tmp", 2:"LinRegress_Sea_tmp"})
projected_sea_tmp_info
```

```python
intermediate_df = global_warming_parameters[["Year", "Sea_anomaly_temp"]]
intermediate_df = intermediate_df[:131]
intermediate_df
```

```python
intermediate_df = intermediate_df.rename(columns = {"Sea_anomaly_temp":"Projected_Sea_tmp"})
intermediate_df_2 = pd.merge(intermediate_df,global_warming_parameters, on = "Year", how = "inner")
intermediate_df_2= intermediate_df_2[["Year", "Projected_Sea_tmp", "Sea_anomaly_temp"]]
intermediate_df_2 = intermediate_df_2.rename(columns = {"Sea_anomaly_temp":"LinRegress_Sea_tmp"})
intermediate_df_2
```

```python
sea_tmp_projections = pd.concat([intermediate_df_2, projected_sea_tmp_info], axis=0)
sea_tmp_projections
```

```python
year_start = 1870
years_reduce = []
for i in range(33):
    year_start = year_start+10
    years_reduce.append(year_start)
years_reduce
```

```python
sea_tmp_projections_reduce = sea_tmp_projections.loc[sea_tmp_projections["Year"].isin(years_reduce),:]
sea_tmp_projections_reduce = sea_tmp_projections_reduce.round(2)
sea_tmp_projections_reduce
```

```python
sea_tmp_projections_reduce.to_csv("Data/Projected_Sea_tmp_reduced.csv", index = False)
```
