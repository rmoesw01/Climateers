```python
import pandas as pd
```

```python
global_warming_parameters = pd.read_csv("Data/All_parameters.csv")
global_warming_parameters.tail(20)
```

```python
global_warming_parameters.columns
```

```python
#showing rows with all data available
global_warming_parameters_reduced = global_warming_parameters[79:131]
```

```python
#predicting sea temp
X_sea_temp=global_warming_parameters_reduced[['GIS_TEMP', 'population','Glacier_mass',
         'CSIRO Adjusted Sea Level']]
y_sea_temp=global_warming_parameters_reduced.Sea_anomaly_temp
```

```python
from sklearn.model_selection import train_test_split
X_train_sea, X_test_sea, y_train_sea, y_test_sea = train_test_split(X_sea_temp, y_sea_temp, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model_sea = LinearRegression()
```

```python
#modeling sea temps
model_sea.fit(X_train_sea, y_train_sea)
training_score_sea = model_sea.score(X_train_sea, y_train_sea)
testing_score_sea = model_sea.score(X_test_sea, y_test_sea)

print(f"Training Score: {training_score_sea}")
print(f"Testing Score: {testing_score_sea}")
```

```python
#get model coefficients

model_sea.coef_
```

```python
projected_values = pd.read_csv("Data/Forecasted_global_warming_values_for_models.csv")
projected_values
```

```python
projected_sea_temps = []
for i in range(len(projected_values["Year"])):
    sea_tmp = (model_sea.coef_[0])*(projected_values["GIS_TEMP"][i]) + (model_sea.coef_[1])*(projected_values["population"][i]) + (model_sea.coef_[2])*(projected_values["Glacier_mass"][i]) + (model_sea.coef_[3])*(projected_values["CSIRO Adjusted Sea Level"][i]) 
    projected_sea_temps.append(sea_tmp)

projected_sea_temps
```
