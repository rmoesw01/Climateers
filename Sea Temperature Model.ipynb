```python
import pandas as pd
```

```python
global_warming_parameters = pd.read_csv("Data/All_parameters.csv")
global_warming_parameters
```

```python
global_warming_parameters.columns
```

```python
#showing rows with all data available
global_warming_parameters_reduced = global_warming_parameters[79:131]
```

```python
#predicting sea temp
X_sea_temp=global_warming_parameters_reduced[['Gas Fuel', 'Liquid Fuel', 'Solid Fuel', 'Cement',
       'Gas Flaring', 'GCAG', 'population',
        'CO2_Increase', 'CSIRO Adjusted Sea Level']]
y_sea_temp=global_warming_parameters_reduced.Sea_anomaly_temp
```

```python
from sklearn.model_selection import train_test_split
X_train_sea, X_test_sea, y_train_sea, y_test_sea = train_test_split(X_sea_temp, y_sea_temp, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model_sea = LinearRegression()
```

```python
#modeling sea temps
model_sea.fit(X_train_sea, y_train_sea)
training_score_sea = model_sea.score(X_train_sea, y_train_sea)
testing_score_sea = model_sea.score(X_test_sea, y_test_sea)

print(f"Training Score: {training_score_sea}")
print(f"Testing Score: {testing_score_sea}")
```
