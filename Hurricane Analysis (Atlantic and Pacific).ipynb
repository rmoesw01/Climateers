```python
import pandas as pd
```

```python
atlantic_hurricanes = pd.read_csv("Data/atlantic_hurricanes.csv")
pacific_hurricanes = pd.read_csv("Data/pacific_hurricanes.csv")
```

```python
# Stack the DataFrames on top of each other
hurricanes_combined = pd.concat([atlantic_hurricanes, pacific_hurricanes], axis=0)
hurricanes_combined
```

```python
#format date
# https://datatofish.com/integers-datetime-pandas-dataframe/
hurricanes_combined['Date'] = pd.to_datetime(hurricanes_combined['Date'], format='%Y%m%d')
```

```python
#extract year and add as a column
hurricanes_combined['Year'] = hurricanes_combined['Date'].dt.year
hurricanes_combined
```

```python
hurricanes_combined["Count"] = 1
hurricanes_combined
```

```python
hurricanes_combined = hurricanes_combined[['Year', 'Status','Count']]
hurricanes_combined
```

```python
years = hurricanes_combined["Year"].unique()
hurricanes = []
tropical_storms = []
tropical_depression = []
for year in years:
    sum_HU = hurricanes_combined.loc[(hurricanes_combined["Status"] == ' HU') & 
                                    (hurricanes_combined["Year"] == year),:]
    hurricanes.append(sum_HU.shape[0])
    sum_TS = hurricanes_combined.loc[(hurricanes_combined["Status"] == ' TS') & 
                                    (hurricanes_combined["Year"] == year),:]
    tropical_storms.append(sum_TS.shape[0])
    sum_TD = hurricanes_combined.loc[(hurricanes_combined["Status"] == ' TD') & 
                                    (hurricanes_combined["Year"] == year),:]
    tropical_depression.append(sum_TD.shape[0])
```

```python
#zip together above lists
all_ocean_data = list(zip(years, hurricanes, tropical_storms,tropical_depression))
#make a dataframe from zipped lists and rename columns
all_ocean_data = pd.DataFrame(all_ocean_data)
all_ocean_data = all_ocean_data.rename(columns = {0:"Year", 1:"Hurricanes", 2:"Tropical_Storms", 3:"Tropical_Depression"})
# atlantic_ocean_data = atlantic_ocean_data[:62]
# atlantic_ocean_data = atlantic_ocean_data[29:160]
all_ocean_data
```

```python
all_parameters = pd.read_csv("Data/All_parameters.csv")
all_parameters
```

```python
gw_data_condensed_hurricanes = pd.merge(all_parameters, all_ocean_data, on = "Year", how = "inner")
gw_data_condensed_hurricanes = gw_data_condensed_hurricanes
gw_data_condensed_hurricanes
```

```python
gw_data_condensed_hurricanes.columns
```

```python
gw_data_condensed_hurricanes_reduced_dates = gw_data_condensed_hurricanes[:130]
```

```python
#tropical_depression analysis
X=gw_data_condensed_hurricanes_reduced_dates[['Gas Fuel', 'Liquid Fuel',  'Cement',
       'population',  'GIS_TEMP',
       'Sea_anomaly_temp',  'CSIRO Adjusted Sea Level'
       ]]
y=gw_data_condensed_hurricanes_reduced_dates.Tropical_Depression
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=32)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#modeling Tropical Depressions
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
#hurricane analysis
X_hurr=gw_data_condensed_hurricanes_reduced_dates[[ 'Liquid Fuel',  'Gas Fuel',
       'Gas Flaring', 'GIS_TEMP',
       'Sea_anomaly_temp', 'GMSL',  'CSIRO Adjusted Sea Level'
       ]]
y_hurr=gw_data_condensed_hurricanes_reduced_dates.Hurricanes
```

```python
X_train_hurr, X_test_hurr, y_train_hurr, y_test_hurr = train_test_split(X_hurr, y_hurr, random_state=12)
```

```python
model_hurr = LinearRegression()
```

```python
#modeling hurricanes
model_hurr.fit(X_train_hurr, y_train_hurr)
training_score_hurr = model_hurr.score(X_train_hurr, y_train_hurr)
testing_score_hurr = model_hurr.score(X_test_hurr, y_test_hurr)

print(f"Training Score: {training_score_hurr}")
print(f"Testing Score: {testing_score_hurr}")
```
